```python
import pandas as pd
```

```python
LABELS_BASE = './Labels/'
blacklist = [
    'Edad',
    'Pais de nacimiento',
    'Total de Personas afrodescendientes',
    'Que Provincia Nació-Nombre'
]

def arreglar_nivel(data, nivel):
    variables = pd.read_csv(LABELS_BASE + f'{nivel}-VARIABLES.csv', sep=';', index_col='variable')
    columnas = variables.index.to_list()
    for columna in columnas:
        try:
            data[columna]
        except KeyError:
            continue

        nombre = variables.loc[columna]['label']
        data.rename(columns={columna: nombre}, inplace=True)
            
        if nombre in blacklist:
            continue
        
        try:
            valores = pd.read_csv(LABELS_BASE + f'{nivel}-{columna}-LABELS.CSV', sep=';')
        except FileNotFoundError as e:
            print(f'Not found! {e.filename}')
            continue

        indices = valores.index.to_list()
        dicc = {indice: valores.loc[indice]['label'] for indice in indices}
        data[nombre] = data[nombre].map(dicc)

    return data
```

```python
data = pd.read_csv('crudos.csv')

# Departamentos y Provincias

deptos = pd.read_csv('DPTO.csv', sep=';', index_col='DPTO')
provs = pd.read_csv('PROV.csv', sep=';', index_col='PROV')

depto_dicc = {depto: deptos.loc[depto]['NOMDPTO'].rstrip() for depto in deptos.index}
prov_dicc = {prov: provs.loc[prov]['NOMPROV'].rstrip() for prov in provs.index}

data['DPTO'] = data['DPTO'].map(depto_dicc)
data['PROV'] = data['PROV'].map(prov_dicc)


# Otras cosas

data = arreglar_nivel(data, 'VIVIENDA')
data = arreglar_nivel(data, 'HOGAR')
data = arreglar_nivel(data, 'PERSONA')

data
```

```python
data.to_csv('cocinado.csv')
```
